```python
import pandas as pd

# Dataframe aanmaken
data = {
    "EmployeeID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Department": ["HR", "IT", "Finance", "Marketing", "Sales"],
    "Salary": [60000, 75000, 80000, 70000, 65000]
}

employees = pd.DataFrame(data)

employees
print()

# Enkel 1ste 3 rijen tonen
print("Enkel 1ste 3 rijen tonen")
print(employees.head(3))
print()

# Summary statistics tonen van de salary-kolom
print("Statistics summary van de salary-kolom")
print(employees["Salary"].describe())
print()

# Aantal employees in elk departement
print("Aantal employees per departement")
print(employees["Department"].value_counts())
print()

# Gemiddelde salaris van de employees
print("Gemiddelde salaris van de employees is", employees["Salary"].mean())
print()

# Employee met het hoogste salaris
max_salaris = employees["Salary"].max()
print("Eployee met het hoogste salaris is", employees[employees["Salary"] == max_salaris])
print()

# Departement met het laagste salarisgemiddelde
min_avg_salaris_dept = employees.groupby("Department")["Salary"].mean().idxmin()
print("Departement met het laagste salarisgemiddelde is", min_avg_salaris_dept)
print()

# Salary increase
employees["Salary increase"] = employees["Salary"] * 1.1
print("Extra kolom met salarisverhoging")
print(employees)
```
